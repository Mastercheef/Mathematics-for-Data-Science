```python
import pandas as pd
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc

%matplotlib inline
```

```python
data = pd.read_json("data\\blogdata.json")
```

```python
w = False
if(w): data=data.transpose()
labels=list(data[0:0][0:len(data)])
```

```python
# Single linkage: Minimale Distanz zwischen Zwei Objekten
# Average linkage: Mittlere Distanz aller Paare von Objekten
# Complete linkage: Maximale Distanz zwischen zwei Objekten 
# Centroid distance: Distanz zwischen den Centroiden der Cluster


plt.figure(figsize=(30,30))
dendrogram = shc.dendrogram(shc.linkage(data.transpose(),method='single',metric="correlation"),orientation="top",labels=labels)
plt.tick_params(axis='x', which='major', labelsize=12)
plt.show()
```
